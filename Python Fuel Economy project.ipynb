import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

```python
mpg = sns.load_dataset("mpg")
mpg.head()
```

```python
mpg.info()
```

## Cleaning up the data for more accruate results


```python
mpg_clean = mpg.dropna()
```

## Relationship Between Horsepower and Fuel Economy


```python
sns.scatterplot( data=mpg_clean,
                 x='horsepower',
y='mpg')
plt.title('Horsepower VS Fuel Economy')
```

### As horsepower goes up, fuel economy generally decreases. Indicating the stronger the horsepower the less the fuel economy becomes.

## Fuel Economy Summary by Country of Origin


```python
mpg_by_origin = mpg_clean.groupby('origin')['mpg'].mean().reset_index().sort_values(by='mpg',ascending=False)
mpg_by_origin.style.format({'mpg': '{:.1f}'})
```

### Vehicles from Japan tend to have the highest average fuel economy.

## Fuel Economy Trends Over Time


```python
sns.lineplot(data=mpg_clean,
         x='model_year',
         y='mpg')
plt.title('Fuel Economy Over Time')
```

### Fuel economy has improved over time with better technology.

## Fuel Economy by Number of Cylinders


```python
sns.barplot(data=mpg_clean,
            x='cylinders',
            y='mpg',
           errorbar=None)
plt.title('Fuel Economy by Number of Cylinders')
```

### Engines with 5 or less cylinders often have better fuel economy then engines with 6 or 8 cylinders

## Fuel Economy by Number of Cylinders


```python
sns.displot(data=mpg_clean, 
            x='mpg', bins=20)
plt.title('Distribution of Fuel Economy')
```

### Most vehicles fall within a mid-range of fuel economy, with fewer vehicles achieving extremely high or low MPG values.


##          Summary


### This analysis shows clear relationships between vehicle characteristics and fuel economy. Higher horsepower and more cylinders are associated with lower MPG, while fuel efficiency has generally improved over time. Differences by country of origin suggest varying design priorities and technology.
